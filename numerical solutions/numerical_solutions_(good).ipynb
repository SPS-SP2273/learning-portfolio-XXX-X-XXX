## Exercise 1 :  The Lorentz Attaractor

```python
import numpy as np
from scipy.integrate import odeint

data_x = [0]
data_y = [1]
data_z = [0]

max_t = 50
initial_t = 0
dt = 0.001
all_t = np.arange(initial_t,max_t,dt)

def dx_dt(x,y,z):
    return (10*(y-x))*dt

def dy_dt(x,y,z):
    return (x*(28-z)-y)*dt

def dz_dt(x,y,z):
    return (x*y-(8/3)*z)*dt

for t in all_t:
    x = data_x[-1]
    y = data_y[-1]
    z = data_z[-1]
    data_x.append(x+dx_dt(x,y,z))
    data_y.append(y+dy_dt(x,y,z))
    data_z.append(z+dz_dt(x,y,z))
```
