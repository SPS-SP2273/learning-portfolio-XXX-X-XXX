### 2.4 Checking for existence

```python
os.mkdir('people')
for person in ['Jonh', 'Paul', 'Ringo']:
    path = os.path.join('people', person)
    if os.path.exists(path):
        print(f'{path} already exists; skipping creation.')
    else:
        os.mkdir(path)
        print(f'Creating {path}')
```

```python
for person in ['Jonh', 'Paul', 'Ringo']:
    path_to_imgs = os.path.join('people', person, 'imgs')
    if not os.path.exists(path_to_imgs):
        os.mkdir(path_to_imgs)

    current_path_of_logo = os.path.join('people', person, 'sp2273_logo.png')
    new_path_of_logo = os.path.join('people', person, 'imgs', 'sp2273_logo.png')

    shutil.move(current_path_of_logo, new_path_of_logo)
    print(f'Moved logo to {new_path_of_logo}')
```

## Exercise 8 :  Tidying a collaboration

```python
import os
import shutil
import glob
```

```python
os.listdir()
```

```python
month = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}
old = 'oct-08_singapore.txt'
new = month[old[:3]]+old[3:]
print(new) #1
```

```python
def rename_my_file(old_file_name): #2
    new_name = month[old_file_name[:3]] + old_file_name[3:]
    return new_name
```

```python
files = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']

for file in files:
    print(file + ' ----> ' + rename_my_file(file)) #3
```

```python
glob.glob('*.txt')
```

```python
for file in glob.glob('*.txt'): #4
    if file[:3] not in month:
        continue
    new_name = rename_my_file(file)
    os.rename(file, new_name)
```

```python
glob.glob('*.txt')
```

```python
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']
for city in cities: #5
    if os.path.exists(city):
        print('Folder already exists!')
        continue
    os.mkdir(city)
```

```python
glob.glob('*cairo.txt')
```

```python
for file in glob.glob('*cairo.txt'): #6
    if file in glob.glob('Cairo/*'):
        continue
    shutil.copy(file, 'Cairo')
```

```python
glob.glob('Cairo/*')
```

```python
for city in cities: #7
    glob_input = f'*{city.lower()}.txt'
    files = glob.glob(glob_input)
    for file in files:
        if file in glob.glob(f'{city}/*'):
            continue
        shutil.copy(file, city)
    print(city + ', complete!')
```
