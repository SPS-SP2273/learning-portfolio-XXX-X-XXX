```python
#3 Housekeeping things
replace_info = {
    'Nanyang Technological University': 'NTU',
    'National University of Singapore': 'NUS',
    'Singapore Management University': 'SMU',
    'Singapore Institute of Technology': 'SIT',
    'Singapore University of Technology and Design': 'SUTD',
    'Singapore University of Social Sciences': 'SUSS'}
new_type_info = {
    'year': 'category',
    'university': 'category',
    'school': 'category',
    'degree': 'category',
    'basic_monthly_median': 'float'
}
columns_to_keep = [
    'year',
    'university',
    'school',
    'degree',
    'basic_monthly_median']
```

```python
data.replace('na', np.nan, inplace=True) #replace na with nan
```

```python
data = data[columns_to_keep] #keeping only basic_monthly_median
```

```python
data.replace(to_replace=replace_info, inplace=True) #shorten uni names
```

```python
data = data.astype(new_type_info) #change types
```

```python
#removing nan
data.dropna(axis=1, inplace=True)
```

```python
#masking for universities to isolate their data
data[data['year'] == 2013]
```
