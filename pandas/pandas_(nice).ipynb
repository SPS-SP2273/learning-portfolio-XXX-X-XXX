```python
#1
data = pn.read_csv('graduate-employment-survey-ntu-nus-sit-smu-suss-sutd.csv', encoding='latin-1')
data
```

```python
#3 Housekeeping things
replace_info = {
    'Nanyang Technological University': 'NTU',
    'National University of Singapore': 'NUS',
    'Singapore Management University': 'SMU',
    'Singapore Institute of Technology': 'SIT',
    'Singapore University of Technology and Design': 'SUTD',
    'Singapore University of Social Sciences': 'SUSS'}
new_type_info = {
    'year': 'category',
    'university': 'category',
    'basic_monthly_median': 'float'
}
columns_to_keep = [
    'year',
    'university',
    'basic_monthly_median']
```

```python
data.replace('na', np.nan, inplace=True) #replace na with nan
```

```python
data = data[columns_to_keep] #keeping only basic_monthly_median
```

```python
data.replace(to_replace=replace_info, inplace=True) #shorten uni names
```

```python
data = data.astype(new_type_info) #change types
```

```python
#removing nan
data.dropna(axis=0, inplace=True)
mask_for_nan = data.isna().any(axis=1)
data[mask_for_nan]
```

```python
#Getting the data needed for Q4
count = data.groupby(by=['year','university']).count().rename(columns={'basic_monthly_median':'count'})
data_min = data.groupby(by=['year','university']).min().rename(columns={'basic_monthly_median':'min'})
data_max = data.groupby(by=['year','university']).max().rename(columns={'basic_monthly_median':'max'})
mean = data.groupby(by=['year','university']).mean().rename(columns={'basic_monthly_median':'mean'})
std = data.groupby(by=['year','university']).std().rename(columns={'basic_monthly_median':'std'})
```

```python
#merging everything together
combined = pn.concat([count,data_min,data_max,mean,std],axis=1)
combined.dropna()
```

```python
data.set_index('year', inplace=True) #I forgot if I used this or if this is important or not
data.set_index('university', append=True, inplace=True)
```

```python
#Generate Pivot Table
data_pivot = pn.pivot_table(data, values='basic_monthly_median', index=['year','university'])
```

```python
#Reset indexing so seaborn can interpret the data for a bar plot
data_graph = data_pivot.reset_index()
```

```python
#Q5 Generating plot w/ data above
sns.set_context('paper')
sns.set_style('dark')
sns.catplot(data=data_graph, x='year', y='basic_monthly_median', hue='university', kind='bar')
```
